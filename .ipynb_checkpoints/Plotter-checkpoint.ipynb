```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from Dice import howMuchWePay 
```

```python
#plot behavior of expected value as we increase the dice throws allowed ==> asymptotic function

diceThrows = np.arange(1,50)
payOff = []
for values in diceThrows:
    payOff.append(howMuchWePay(values,10,5))
```

```python
plt.plot(diceThrows, payOff)
```

```python
diceThrows = np.arange(1,15)
payOff = []
for values in diceThrows:
    payOff.append(howMuchWePay(values,6))

```

```python
plt.plot(diceThrows, payOff)
```
